```python
num = input()
```

```python
num = num.split(" ")
num
```

```python
score = []
for i in range(len(num)-1):
    if (i+1) % 11 == 7:
        score.append(num[i])
```

```python
score_values = []
for j in score:
    # st = st[:-1]
    u = j[:-1]
    score_values.append(float(u))
score_values
```

```python
# PLOT THE SCORES
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

score = score_values[:224]
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(score)), score)
plt.ylabel('Score')
plt.xlabel('Episode Number')
plt.show()
```
